```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_excel('/content/drive/My Drive/Colab Notebooks/data.xlsx')
data.head()
```

```python
data.groupby(by='WORKCLASS')['HOURS PER WEEK'].mean()
```

```python
df = data[data.OCCUPATION !='?']
```

```python
df.loc[df['COUNTRY']!='United-States','COUNTRY'] = 'non_usa'
```

```python
for i in df.columns:
    if type(df[i][1])== str:
        print(df[i].value_counts())
```

```python
df.columns
```

```python
import seaborn as sns
fig, ((a,b),(c,d),(e,f)) = plt.subplots(3,2,figsize=(20,25))
plt.xticks(rotation=45)
sns.countplot(df['WORKCLASS'],hue=df['CLASS'],ax=f)
sns.countplot(df['RELATIONSHIP'],hue=df['CLASS'],ax=b)
sns.countplot(df['MARTIAL_STATUS'],hue=df['CLASS'],ax=c)
sns.countplot(df['AGE'],hue=df['CLASS'],ax=d)
sns.countplot(df['GENDER'],hue=df['CLASS'],ax=e)
sns.countplot(df['COUNTRY'],hue=df['CLASS'],ax=a)
```

```python
plt.figure(figsize=(24, 6))
ro = sns.countplot(df['OCCUPATION'], hue=df['GENDER'])
ro.set_xticklabels(ro.get_xticklabels(), rotation=30, ha="right")
ro.get_figure().savefig("/content/drive/My Drive/Colab Notebooks/GEN_VS_OCCUPATION.png")
plt.show()
```
