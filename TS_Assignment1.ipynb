Assignment 1.1: Python program to find the largest element among three Numbers

```
USER_INPUT=list(map(int,input("Please Enter Numbers separted from commas: ").split(",")))
minimum = maximum = None;
for num in USER_INPUT:
    if(minimum == None):
        minimum = num
    if(maximum == None):
        maximum  = num
    if(num<minimum):
        minimum = num
    if(num>maximum):
        maximum = num
print ("Array: ", USER_INPUT)
print("Minimum No:{0} , Maximum No:{1}".format(minimum,maximum))
```

Assignment 1.2: Python Program to check given number is Prime number or not

```
num = int (input("Please provide the number for prime check:"))
if num>1:
    temp = num**0.5+1
    index = 2
    while (index<temp):
        if(num!=index and num%index==0):
            print("{number} is not prime".format(number=str(num)))
            break
        index+=1
    else:
        print("{number} is prime".format(number=str(num)))
else:
    print("{number} is not prime".format(number=str(num)))

    
```

Assignment 1.3: Python Program to display all prime numbers within an interval

```
start = int(input("Please Enter Starting Number:"))
end   = int(input("Please Enter Ending Number:"))

primeList = []

for num in range(start,end+1):
    if num>1:
        temp = num**0.5+1
        index = 2
        while (index<=temp):
            if(num!=index and num%index==0):
                #print("{number} is not prime".format(number=str(num)))
                break
            index+=1
        else:
            primeList.append(num)
            #print("{number} is prime".format(number=str(num)))
    else:
        print("{number} is not prime".format(number=str(num)))

print ("Prime Numbers from {startnum} to {endnum}: {listnum}".format(startnum=start,endnum=end,listnum=primeList))
```

Assignment 2.1: Python Program to Check where a String is Palindrome or not ?

```
user_input = input("Please Enter the String:").lower()
length = len(user_input)

for index in range(0,int(length/2)):
    if(user_input[index]!=user_input[length-index-1]):
        print("{input} is not a palindrome.".format(input=user_input))
        break;
else:
    print("{input} is a palindrome.".format(input=user_input))
```

Assignment 3.1: Python program to print Highest Common Factor (HCF) of two numbers

```
def HCF(num1,num2):
    if(num2==0):
        return num1
    else:
        return HCF(num2,num1%num2)
        
num1 = int(input("Please Enetr First Number: "))
num2 = int(input("Please Enetr Second Number: "))

print("HCF of {number1} and {number2} : {HCF}".format(number1=num1,number2=num2,HCF= HCF(num1,num2)))
```

Assignment 3.2: Python program to make a simple calculator that can add, subtract, multiply and division

```
def sum_numbers(a, b):
    return a+b

def substract_numbers(a, b):
    return a-b

def multiply_numbers(a, b):
    return a*b

def divide_numbers(a, b):
    return a/b

while True:
    operation = int(input("Please Select Operation Number:\n 1.Add \n 2.Substract \n 3.Multiply \n 4.Divide \n 5.Exit "))
    num1 = int(input("Please Enter First Number:"))
    num2 = int(input("Please Enter Second Number:"))
    if(operation==1):
        print("{a}+{b}= {add}".format(a=num1,b=num2,add=sum_numbers(num1,num2)))
    if(operation==2):
        print("{a}-{b}= {add}".format(a=num1,b=num2,add=substract_numbers(num1,num2)))
    if(operation==3):
        print("{a}+{b}= {add}".format(a=num1,b=num2,add=multiply_numbers(num1,num2)))
    if(operation==4):
        print("{a}+{b}= {add}".format(a=num1,b=num2,add=divide_numbers(num1,num2)))
    if(operation==5):
        break;
```

Assignment 3.3: Python program to display the fibonacci sequence up to n-th term using recursive function

```
def fibonnaci_num(nterm):
  if nterm <= 1:
    return 1
  else:
    return fibonnaci_num(nterm-1)+fibonnaci_num(nterm-2)
    
def fib_series():
  list = []
  term = int(input("Enter length of Fibonacci series:"))
  for i in range(term):
    list.append(fibonnaci_num(i))
  print(list)
  
fib_series()
```

**K MEANS**

```
Numeric_data = custdata[['Fresh','Milk','Grocery','Frozen','Detergents','Delicassen','Clothing','Stationary']]
Numeric_data.columns
```

```
Numeric_data['Stationary'].fillna(Numeric_data['Stationary'].median(),inplace=True)
Numeric_data.head()
```

```
#This estimator scales and transform the data into zero to one range
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(Numeric_data)
scaled_data = pd.DataFrame(scaled_data,columns=Numeric_data.columns)
scaled_data.head()
```

Finding no of clusters

```
X1 = scaled_data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents', 'Delicassen','Clothing']].iloc[: , :].values
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```
plt.figure(1 , figsize = (6 ,6))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```
model = KMeans(n_clusters=4,random_state=7)
model.fit(scaled_data)
```

```
clusters_data = customer_df.copy()
clusters_data.head()
```

```
clusters_data['clusters']=model.predict(scaled_data)
clusters_data.head()
```

```
clust_prof = clusters_data.groupby(['clusters'],as_index=False).mean()
clust_prof
```

```
cluster_deviation=[]
cluster_zscore=[]

for x in ['Fresh', 'Milk', 'Grocery','Detergents','Frozen','Delicassen']:
    devi=clust_prof[x]-clust_mean[x]
    cluster_deviation.append(devi)
    zsc=devi/clust_std[x]
    cluster_zscore.append(zsc)
```
